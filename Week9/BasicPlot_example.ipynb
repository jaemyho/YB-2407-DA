```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.plot(x, y, marker='o', color='red', linestyle='--', linewidth=2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Baisc Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['A','B','C','D']
values = [3,7,1,8]
plt.bar(categories,values,color='green', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for multiple datasets
np.random.seed(0)  # For reproducibility
x1 = np.random.rand(50) * 10  # 50 random points for x1
y1 = np.random.rand(50) * 10  # 50 random points for y1
x2 = np.random.rand(50) * 10  # 50 random points for x2
y2 = np.random.rand(50) * 10  # 50 random points for y2

# Scatter plot with different markers and colors
plt.scatter(x1, y1, color='blue', label='Dataset 1', marker='o', alpha=0.6)
plt.scatter(x2, y2, color='red', label='Dataset 2', marker='^', alpha=0.6)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Complex Scatter Plot with Multiple Datasets')

plt.legend()

plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
sizes = [30, 40, 20, 10]
colors = ['red', 'yellow', 'pink', 'purple'] 

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Fruit Distribution')
plt.show()
```

```python
data = [
    [5, 7, 8, 6, 7, 9, 8, 6, 5, 7],  # Group 1
    [15, 16, 14, 18, 17, 19, 16, 15, 14, 17],  # Group 2
    [25, 22, 24, 23, 26, 27, 23, 25, 22, 24]   # Group 3
]

plt.boxplot(data, vert=True, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='black'),
            flierprops=dict(marker='o', color='red', markersize=5))

plt.xlabel('Group')
plt.ylabel('Value')
plt.title('Box Plot with Specific Data')
plt.show()
```
