```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the uploaded datasets
file_paths = [
    './Iris/bezdekIris.csv',
    './Iris/irisdata.csv'
]
for file_path in file_paths:
    try:
        df = pd.read_csv(file_path, header=None)
        print(f"Successfully loaded: {file_path}")
        print(df.head())  # Display first few rows for inspection
    except Exception as e:
        print(f"Error loading {file_path}: {e}")

# Assign column names based on the structure of the Iris dataset
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Initial Data Cleaning
# 1. Check for missing values
missing_values = df.isnull().sum()
print("Missing Values : ")
print(missing_values)

# 2. Rate of missing values
missing_rates = df.isnull().sum() / len(df)
print("Missing Rates : ")
print(missing_rates)

# 3. Drop duplicates
df_cleaned = df.drop_duplicates()
print(df_cleaned)


# Calculate the correlation matrix for the numerical data
correlation_matrix = df_cleaned.corr(numeric_only=True)
print(correlation_matrix)


# Heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()

```
